# Lab Events Dataset

```python
lab_events.head()

# drop row_id
lab_events = lab_events.drop(columns={'ROW_ID'})
lab_events.head()
```

```python
# drop rows where HADM is null 
lab_events = lab_events.drop(lab_events[lab_events.HADM_ID.isnull()].index)
lab_events.head()
```

```python
lab_events.info()
```

```python
lab_events['SUBJECT_ID'].nunique()
```

```python
lab_events['HADM_ID'].nunique()

# convert HADM_ID into int 
lab_events['HADM_ID'] = lab_events['HADM_ID'].astype(int)
lab_events.info()
```

```python
lab_events_short = lab_events[['SUBJECT_ID', 'HADM_ID', 'ITEMID', 'VALUENUM', 'VALUEUOM']]
# interested in body temperature (50825) & blood oxygen saturation level (50817)
lab_events_short = lab_events_short[(lab_events_short['ITEMID'] == 50825) | (lab_events_short['ITEMID'] == 50817)]
# getting average body temperature & blood oxygen saturation level 
lab_events_short['AVERAGE'] = lab_events_short.groupby(['SUBJECT_ID', 'HADM_ID', 'ITEMID'])['VALUENUM'].transform('mean')
lab_events_short = lab_events_short[['SUBJECT_ID', 'HADM_ID', 'ITEMID', 'AVERAGE', 'VALUEUOM']]
lab_events_short = lab_events_short.drop_duplicates()
lab_events_short
```

# Chart Event Dataset

```python
chart_event.info()
```

```python
chart_event_short = chart_event[['SUBJECT_ID', 'HADM_ID', 'ITEMID', 'CHARTTIME', 'STORETIME', 'VALUE', 'VALUEUOM']]
# getting average heart rate bpm - group by subject_id & hadm_id
chart_event_short['AVERAGE'] = chart_event_short.groupby(['SUBJECT_ID', 'HADM_ID', 'ITEMID'])['VALUE'].transform('mean')
chart_event_short = chart_event_short[['SUBJECT_ID', 'HADM_ID', 'ITEMID', 'AVERAGE', 'VALUEUOM']]
chart_event_short = chart_event_short.drop_duplicates()
chart_event_short
```

# Complete Dataset

```python
df = pd.merge(admissions_short, patients_short, left_on = 'SUBJECT_ID', right_on = 'SUBJECT_ID', how = 'left')
df.head()
```

```python
chart_event_short = chart_event_short.rename(columns={'AVERAGE':'VALUE'})
lab_events_short = lab_events_short.rename(columns={'AVERAGE':'VALUE'})
vital_signs = lab_events_short.append(chart_event_short, ignore_index=True)
vital_signs
```

```python
df = pd.merge(df, vital_signs, how = 'left', left_on = ['SUBJECT_ID', 'HADM_ID'], right_on = ['SUBJECT_ID', 'HADM_ID'])
df.head()
```

```python
lab_items_short = lab_items_short[(lab_items_short['ITEMID'] == 50825) | (lab_items_short['ITEMID'] == 50817)]
items_short = items_short[(items_short['ITEMID'] == 220045)]
items_total = lab_items_short.append(items_short, ignore_index=True)
items_total
```

```python
df = pd.merge(df, items_total, how = 'left', left_on = ['ITEMID'], right_on = ['ITEMID'])
df.head()
```

```python
df.shape
```

# Data Cleaning

```python
# remove rows where ITEMID is null

df = df[df['ITEMID'].notna()]
df
```

```python
# removing duplicates
df = df.drop_duplicates()
df
```
